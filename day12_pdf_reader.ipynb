```python
import PyPDF2
import anthropic
from google.colab import userdata
print("all tools loaded")
```

```python
api_key = userdata.get('ANTHROPIC_KEY')
client = anthropic.Anthropic(api_key=api_key)
print("connected to claude")
```

```python
# open and read the pdf file
pdf_file = open('Profilej.pdf', 'rb')
pdf_reader = PyPDF2.PdfReader(pdf_file)

#Count how many pages
total_pages = len(pdf_reader.pages)
print(f"PDF loaded! Total pages: {total_pages}")

#Extract text from every page
document_text = ""
for page_num in range(total_pages):
  page = pdf_reader.pages[page_num]
  document_text += page.extract_text()

print(f"Text extracted! Total characters: {len(document_text)}")
print("\nFirst 500 characters of the PDF:")
print(document_text[:500])

```

```python
# Ask Claude a question about the PDF content
question = "What is this document about and what did the person achieve?"

prompt = f"""
You are a document assistant. Read this document carefully and answer the question.

DOCUMENT CONTENT:
{document_text}

QUESTION:
{question}

Give a clear and helpful answer based only on the document.
"""

message = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=1024,
    messages=[{"role": "user", "content": prompt}]
)

print("=== Claude's Answer ===")
print(message.content[0].text)
```

```python
print("=== PDF Question Answering System ===")
print(f"Document loaded: Profilej.pdf ({total_pages} pages)")
print("Type your questions below. Type 'quit' to stop.")
print("---")

while True:
    question = input("\nYour question: ")

    if question.lower() == 'quit':
        print("Goodbye!")
        break


    if question.strip() == "":
        print("Please type a question!")
        continue

    prompt = f"""
You are a document assistant. Answer based ONLY on the document provided.
If the answer is not in the document, say "This information is not in the document."

DOCUMENT:
{document_text}

QUESTION:
{question}

Give a short and direct answer.
"""

    message = client.messages.create(
        model="claude-haiku-4-5-20251001",
        max_tokens=512,
        messages=[{"role": "user", "content": prompt}]
    )

    print(f"\nAnswer: {message.content[0].text}")
    print("---")
```
