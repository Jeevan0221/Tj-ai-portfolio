```python
import anthropic
from google.colab import userdata

#Load API Key safely from colab secrets
client = anthropic.Anthropic(api_key=userdata.get('ANTHROPIC_KEY'))

def ai_writing_assistant(task, text):
    prompts = {
        "write_email":f"write a professional email about: {text}",
        "summarize":f"summarize this text in 3 bullet points: {text}",
        "fix_grammar":f"Fix all grammar and spelling errors, return only the corrected version:{text}",
        "make_formal":f"Rewrite this in a formal professional tone: {text}",
        "make_casual":f"Rewrite this in a friendly casual tone: {text}"
    }

    response = client.messages.create(
        model="claude-haiku-4-5-20251001",
        max_tokens=500,
        messages=[{"role":"user", "content": prompts[task]}]
    )
    return response.content[0].text

#---- Test It-----
print("=== Write an email ===")
print(ai_writing_assistant("write_email", "asking my manager for a day off next Friday"))

print("\n=== SUMMARIZE ===")
print(ai_writing_assistant("summarize",
    "AI engineers are in high demand in 2026. Companies are using AI to automate tasks and improve products. Salaries are growing every year and learning AI skills now puts you ahead of most job seekers."))

print("\n=== FIX GRAMMAR ===")
print(ai_writing_assistant("fix_grammar",
    "i wants to learn ai and get a job in technology field as soon as possible"))

print("\n=== MAKE FORMAL ===")
print(ai_writing_assistant("make_formal",
    "hey can u check my resume and tell me if its good enough"))

print("\n=== MAKE CASUAL ===")
print(ai_writing_assistant("make_casual",
    "I would like to formally request your assistance in reviewing my professional portfolio"))
```
