```python
#Lists -Store multiple items
skills =["Python", "AI", "Prompt Engineering", "Machine Learning"]
print("My skills:", skills)
```

```python
#Loops - repeat actions automatically
print("\n--- My AI Skills ---")
for skill in skills:
  print(" I am Learning:", skill)
```

```python
#FUNCTIONS - reusable blocks of code
def filter_ai_jobs(job_list):
  ai_jobs = []
  for job in job_list:
    if "AI" in job or "Machine Learning" in job or "Data" in job:
      ai_jobs.append(job)
  return ai_jobs
```

```python
#Test your function
jobs = ["AI Engineer", "Web Developer", "Machine Learning Engineer",
        "Graphic Designer", "Data Scientist", "Accountant"]
my_jobs = filter_ai_jobs(jobs)
print("\n--- Jobs I'm Targeting ---")
for job in my_jobs:
     print("✅", job)
```
