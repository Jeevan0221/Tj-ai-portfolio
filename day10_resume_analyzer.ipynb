```python
import anthropic
from google.colab import userdata

print("Tools loaded")
```

```python
#Connect to claude
api_key = userdata.get('ANTHROPIC_KEY')
client = anthropic.Anthropic(api_key=api_key)

print("Connected to claude!")
```

```python
#Here we paste a sample job description and sample resume

job_description = """
We are looking for an AI Engineer with experience in:
- Python programming
- Machine Learning and scikit-learn
- Large Language Models (LLMs) and Claude API
- AWS Cloud services (S3, EC2, Lambda)
- Data analysis and pandas
- Prompt Engineering
- Building AI-powered applications
"""

my_resume = """
Senior Data Engineer with 8 years experience at Bank of America, Verizon, and Nielsen.
Skills: Python, SQL, AWS (S3, EC2, Lambda), data pipelines, pandas
Currently learning: Claude API, Machine Learning, scikit-learn, Prompt Engineering
Projects: Built AI chatbot, ML salary predictor, AI data analyst using Claude API
"""
```

```python
#Ask claude to analyze

prompt=f"""
You are an expert resume reviewer and career coach

Analyze this resume against the job description and provide feedback
1. Match score out of 100
2. Top 5 keywords from job description that are missing or weak in the resume
3. Top 3 specific improvements to make the resume stronger
4. One  honest sentence about the candidates chances

JOB DESCRIPTION:
{job_description}

RESUME:
{my_resume}

Be specific, Honest and Helpful.
"""

message = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=1024,
    messages=[
        {"role":"user","content": prompt}
    ]
)

print("===AI Resume Analysis===")
print(message.content[0].text)
```
