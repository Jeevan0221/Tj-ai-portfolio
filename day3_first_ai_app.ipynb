```python
import anthropic

# connect to claude AI
# Replace "Your_api_key" with your actual Anthropic API key
client = anthropic.Anthropic(api_key="YOUR_ANTHROPIC_API_KEY_HERE")

# Function that asks AI anything
def ask_ai(question):
  response = client.messages.create(
      model="claude-haiku-4-5-20251001",
      max_tokens=500,
      messages=[{"role":"user","content":question}]
  )
  return response.content[0].text

# Test it with 3 questions
print("Question 1:")
print(ask_ai("What are the top 3 skills to get an AI job in 2026?"))

print("Question 2:")
print(ask_ai("Write one line python tip for beginners"))

print("Question 3:")
print(ask_ai("Give me a motivavtional quote about learning AI"))
```
