```python
import anthropic
from google.colab import userdata

#Safely loads your API key from colab secrets
client = anthropic.Anthropic(api_key=userdata.get('ANTHROPIC_KEY'))

#Memory to store conversation history
conversation_history = []

def chat_with_ai(user_message):
    # Add your message to history
    conversation_history.append({
        "role": "user",
        "content": user_message
    })

    # Send to claude AI
    response = client.messages.create(
        model="claude-haiku-4-5-20251001",
        max_tokens=500,
        system="You are TJ's personal AI assistant helping him learn AI engineering and get a job in tech",
        messages=conversation_history
    )

    #Get AI response
    ai_response = response.content[0].text

    #save AI response to history
    conversation_history.append({
        "role": "assistant",
        "content": ai_response
    })

    return ai_response

#Start Chatting!
print("🤖 TJ's Personal AI Assistant")
print("Type your question and press Enter.")
print("Type 'quit' to stop.\n")

while True:
    user_input = input("You:")
    if user_input.lower() == "quit":
        print("Goodbye!")
        break
    response = chat_with_ai(user_input)
    print(f"AI: {response}")
```
