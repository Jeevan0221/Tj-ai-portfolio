```python
import anthropic
from google.colab import userdata

print("Tools loaded")
```

```python
api_key = userdata.get('ANTHROPIC_KEY')
client = anthropic.Anthropic(api_key=api_key)

print("connected to claude")
```

```python
# This is our company document -stored as a text variable

document = """
COMPANY LEAVE POLICY - 2026

1. TYPES OF LEAVE
- Annual Leave: Every employee gets 20 days per year
- Sick Leave: Every employee gets 10 days per year
- Emergency Leave: 3 days per year for personal emergencies
- Maternity Leave: 90 days for female employees
- Paternity Leave: 10 days for male employees

2. LEAVE APPROVAL PROCESS
- All leave requests must be submitted at least 3 days in advance
- Annual leave requires approval from direct manager
- Sick leave requires a medical certificate if more than 2 consecutive days
- Emergency leaves must be communicated to manager within 24 hours
- Maternity and Paternity leave requires HR approval 30 days in advance

3. LEAVE RULES
- Annual leave cannot be carried forward to next year
- Maximum 5 consecutive days of annual leave at one time
- Leave during project deadline requires special approval from department head
- Unused sick leave is not paid out at end of year
- New employees must complete 6 months before taking annual leave
"""

print("Document loaded!")
print(f"Document length: {len(document)} characters")
```

```python
# Your 3 questions from earlier

questions = """
1. What are the policies applicable for employees to take leave
2. Does leave require any approval from the manager?
3. How many days of leave does each employee get?"""

# Build the prompt - documnt + questions together
prompt = f"""
You are a Strict lawyer. A manager has given you a company leave policy document
and has 3 questions.

Read the document carefully and answer all 3 questions clearly.

COMPANY DOCUMENT:{document}
MANAGER'S QUESTIONS:{questions}

Answer each question separately with a clear heading.questions.
"""

# Send to claude
message = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=1024,
    messages=[{"role": "user", "content": prompt}]
)

print("===HR Assistant response===")
print(message.content[0].text)

```

```python
questions_list = [
    "How many annual leave days do employees get?",
    "What happens to unused sick leave at end of year?",
    "Can a new employee take leave in their first month?"
]

print("=== Document Q&A System ===")
print(f"Document loaded. Ready to answer {len(questions_list)} questions")
print("----")

# Ask each question seperately in a loop
for i, question in enumerate(questions_list):

    prompt = f"""
You are an HR assistant. Answer this question based ONLY on the document provided.
If the answer is not in the document, say "This information is not in the document."

DOCUMENT:{document}

QUESTION:{question}

Give a short and direct answer in 2-3 sentences maximum.
"""

    message = client.messages.create(
      model="claude-haiku-4-5-20251001",
      max_tokens=256,
      messages=[{"role": "user", "content": prompt}]
    )

    print(f"Question {i+1}: {question}")
    print(f"Answer: {message.content[0].text}")
    print("----")
```

```python
prompt = f"""
You are a professional document analyst.

Read the following company document carefully and provide:

1. SUMMARY - A 3 sentence overview of what this document is about
2. KEY FACTS - The 5 most important facts from the document as bullet points
3. IMPORTANT RULES - Any rules employees must follow, listed clearly
4. RED FLAGS - Anything in this document that employees should pay special attention to
5. QUESTIONS TO ASK HR - 3 questions an employee should ask HR after reading this document

DOCUMENT:
{document}

Be concise and professional.
"""

message = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=1024,
    messages=[{"role": "user", "content": prompt}]
)

print("=== Automatic Document Analysis ===")
print(message.content[0].text)
```
